```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
answer_dataset = pd.read_csv("Dataset/answerdatacorrect.csv")
subject_metadata = pd.read_csv("Dataset/subject_metadata.csv")
```

# Duplicates count

```python
print('No of duplicates in answerdata: {}'.format(sum(answer_dataset.duplicated())))
print('No of duplicates in subject_metadata: {}'.format(sum(subject_metadata.duplicated())))
```

# Null vals

```python
print('We have {} NaN/Null values in answerdata'.format(answer_dataset.isnull().values.sum()))
print('We have {} NaN/Null values in subject_metadata'.format(subject_metadata.isnull().values.sum()))
```

# answerdata

```python
cols = [col for col in answer_dataset.columns if col != "DateOfBirth" and col != "PremiumPupil" and col != "DateAnswered" and col != "SubjectId" 
        and col !="Region" and col != "CountryCode"]

fig, ax = plt.subplots(1, len(cols), figsize=(25, 10))
for i, c in enumerate(cols):
    ax[i].hist(answer_dataset[c], edgecolor='white')
    ax[i].set_xlabel(c, fontsize=10)

plt.show()
```

```python
cols = [col for col in answer_dataset.columns if col != "DateOfBirth" and col != "PremiumPupil" and col != "DateAnswered" and col != "SubjectId" 
        and col !="Region" and col != "CountryCode"]

fig, ax = plt.subplots(1,len(cols), figsize=(25, 10))

for i, c in enumerate(cols):
    ax[i].boxplot(answer_dataset[c])
    
    ax[i].set_xticks([])
    ax[i].set_xlabel(c, fontsize=10)
```

```python
cols = [col for col in answer_dataset.columns if col != "DateOfBirth" and col != "PremiumPupil" and col != "DateAnswered" and col != "SubjectId" 
        and col !="Region" and col != "CountryCode"]

fig, ax = plt.subplots(1,len(cols), figsize=(25, 10))

for i, c in enumerate(cols):
    ax[i].violinplot(answer_dataset[c])
    
    ax[i].set_xticks([])
    ax[i].set_xlabel(c, fontsize=10)
```

```python
plt.figure(figsize=(15,10))
a = answer_dataset.corr()
# Generate a mask for upper traingle
mask = np.triu(np.ones_like(a, dtype=bool))
# Configure a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)
# Draw the heatmap
sns.heatmap(a, annot=True, mask = mask, cmap=cmap)
plt.show()
```
